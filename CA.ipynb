```python
import http.client
import urllib.request

http.client._MAXHEADERS = 1000

url = "https://kworb.net/itunes/"
response = urllib.request.urlopen(url)
raw_html = response.read().decode("utf-8")
soup = BeautifulSoup(raw_html, 'html.parser')
print(soup)
```

```python
def getInfo(trnode): #####how it changed frm ulnode to trnode
    print("Artist Name:", trnode.find_all("a")[0].get_text())
    country_cell = trnode.find("td", class_="bl mp smaller") #####why store in variable?
    print("Top Country:", country_cell.get_text())### why no index

ulnode = soup.find("table", {"id":"artists"}).find("tbody")
for linode in ulnode.find_all("tr"):### why tr
    #print(linode.get_text())
    getInfo(linode)
    print()
```

```python
def getInfoData(trnode):
    datarow = {}
    datarow["Artist Name"] = trnode.find_all("a")[0].get_text()
    datarow["Top Country"] = trnode.find("td", class_="bl mp smaller").get_text()
    return datarow

response = urllib.request.urlopen(url)
raw_html = response.read().decode("utf-8")
soup = BeautifulSoup(raw_html, 'html.parser')
ulnode = soup.find("table", {"id":"artists"}).find("tbody")
country_items = []
for linode in ulnode.find_all("tr"):
    country_items.append(getInfoData(linode))
for item in country_items[:5]:
    print(item)
```

```python
import pprint
pprint.pprint(country_items)
```

```python
from prettytable import PrettyTable
xTable = PrettyTable()
xTable.field_names = ["Artist Name", "Top Country"]
xTable.align["Artist Name"] = xTable.align["Top Country"] = "l"

for x in country_items:
    xTable.add_row([x["Artist Name"], x["Top Country"]])
print(xTable)
```
